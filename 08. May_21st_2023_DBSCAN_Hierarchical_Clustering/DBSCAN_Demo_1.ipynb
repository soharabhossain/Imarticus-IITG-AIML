# Compare K-means and DBSCAN Algorithms

## Import the libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

## Read and Visualize the Data

## Read the Blobs Data

```python
# Read data in dataframes

blobs = pd.read_csv('cluster_blobs.csv')

blobs.head()
```

```python
sns.scatterplot(data=blobs, x='X1', y='X2');
```

## Read the Mooons

```python
moons = pd.read_csv('cluster_moons.csv')

moons.head()
```

```python
sns.scatterplot(data=moons, x='X1', y='X2');
```

## Read the Circles Data

```python
circles = pd.read_csv('cluster_circles.csv')

circles.head()
```

```python
sns.scatterplot(data=circles, x='X1', y='X2');
```

## Write a Custom function to apply the desired clustering method and plot the clusters

```python
def display_categories (model, data):
  labels = model.fit_predict(data)
  sns.scatterplot(data=data, x='X1', y='X2', hue=labels, palette='Set1');
```

# Apply K-means Clustering

```python
model = KMeans(n_clusters=3, n_init='auto')
```

## Apply K-means on Blobs Data

```python
display_categories(model, blobs)
```

## Apply K-means on Moons Data

```python
kmeans = KMeans(n_clusters=2, n_init='auto')

display_categories(kmeans, moons)
```

## Apply K-means on Circles Data

```python
display_categories(model, circles)
```

```python
display_categories(kmeans, circles)
```

# Apply DBSCAN Clustering

```python
model = DBSCAN()
```

## Apply DBSCAN on Blobs Data

```python
display_categories(model, blobs)
```

```python
model = DBSCAN(eps=1.5)
display_categories(model, blobs)
```

## Apply DBSCAN on Moons Dataset

```python
#model = DBSCAN()
model = DBSCAN(eps=0.15)


display_categories(model, moons)
```

## Apply DBSCAN to Circles Data

```python
#model = DBSCAN()
model = DBSCAN(eps=0.15)
display_categories(model, circles)
```
