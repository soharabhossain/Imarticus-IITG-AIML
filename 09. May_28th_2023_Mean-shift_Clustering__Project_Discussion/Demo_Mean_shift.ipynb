# Mean-shift Algorithm Coding Demo

## Import necessary Libraries

```python
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

## Generate Dta Points

```python
data_points, _ = make_blobs(n_samples=200, centers=2, random_state=23) # reproducability of the results
#data_points, _ = make_blobs(n_samples=200, centers=2)
```

```python
data_points.shape
```

## Apply Mean-shift Clustering

```python
ms = MeanShift()
```

```python
ms.fit(data_points)
```

```python
labels = ms.labels_
labels
```

```python
cluster_centers = ms.cluster_centers_
cluster_centers
```

## Plot the Data

```python
plt.scatter(data_points[:,0], data_points[:,1], c=labels)
plt.scatter(cluster_centers[:,0], cluster_centers[:,1], marker='x', color='red')
plt.title('Mean-shift Clustering')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Apply Mean-shift on another Dataset

```python

from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.datasets import make_blobs
```

```python
centers = [[1,1], [-1,-1], [1,-1]]
X , _ = make_blobs(n_samples=10000, centers=centers, cluster_std=0.6)
```

```python
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=500)
bandwidth
```

```python
#ms = MeanShift(bandwidth=bandwidth, bin_seeding=True) # computationally less exensive

ms = MeanShift(bandwidth=bandwidth) # without bin-seeding computationally more exensive

ms.fit(X)
```

```python
labels = ms.labels_
labels
```

```python
cluster_centers = ms.cluster_centers_
cluster_centers
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(cluster_centers[:,0], cluster_centers[:,1], marker='x', color='red')
plt.title('Mean-shift Clustering')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
