# Mean-shift for Image Color Segmentation

## Import the Libraries

```python
import numpy as np
from sklearn.cluster import MeanShift, estimate_bandwidth
from PIL import Image
import matplotlib.pyplot as plt
```

## Read the Image

```python
#image = Image.open('dog.jpg')
image = Image.open('tiger.jpg')
#image = Image.open('peeper.jpg')
```

```python
image = np.array(image)
```

```python
shape = image.shape
shape
```

```python
reshape_img = np.reshape(image, [-1, 3])
```

```python
plt.imshow(image)
plt.title('Input Image')
```

## Apply Mean-shift

```python
bandwidth = estimate_bandwidth(reshape_img, quantile=0.2, n_samples=100)
```

```python
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(reshape_img)
```

```python
print('Number of estimated clusters: %d' % len(np.unique(ms.labels_)))
```

```python
labels = ms.labels_
result_image = np.reshape(labels, shape[:2]) 
```

```python
fig = plt.figure(2, figsize=(14, 12))

ax = fig.add_subplot(121)
ax = plt.imshow(image)

ax = fig.add_subplot(122)
ax = plt.imshow(result_image)
plt.show()
```
